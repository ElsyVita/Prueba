```python
#@title **TEST**
#@markdown Enter the number of subprocesses to split.
time = 4 #@param {type:"number"}
time1 = 801 #@param {type:"slider", min:1, max:2000, step:1}
name_folder = 'data' #@param {type:"string"}


from matplotlib import pyplot as plt
import numpy as np
import scipy.signal as signal
import math
import os

path_folder = '/content/drive/MyDrive/' + name_folder + '/'
if not os.path.exists(path_folder):
    os.makedirs(path_folder)

# Define Transfer Function
num = np.array([2])
den = np.array([3 , 1])
H = signal.TransferFunction(num , den)
print ('H(s) =', H)
# Step Response
x, y = signal.step(H)

#x = []
#y = []
#for i in range(0, time1, 1):
    #xi = i/100 + 1
    # yi = xi*xi
    #yi = math.log(xi)
    #x.append(xi)
    #y.append(yi)

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('t')
# naming the y axis
plt.ylabel('y')

# giving a title to my graph
plt.title('Step Response')

plt.grid()

name_imagen= path_folder + 'fig1.png'
plt.savefig(name_imagen)

# function to show the plot
plt.show()
```

```python
#@title **Save results**
#@markdown Save results in google drive.
name_file = 'data_simul' #@param {type:"string"}

import pandas


data={"X": x,
      "Y": y}

path_file= path_folder + name_file + '.csv'

#path_file = '/content/drive/MyDrive/' + name_file + '.csv'
df = pandas.DataFrame(data)
df.to_csv(path_file, sep=',',index=False)
```
